# Imputation : 
   1. Mean
   2. Median
   3. Mode

```python
# Mean
ageMean = df['Age'].mean()
```

```python
# Median
ageMedian = df['Age'].median()
ageMedian
```

```python
# Mode
ageMode = df['Age'].mode()
ageMode
```

## Filling the null values with 'Mean'

```python
df['Age'].fillna(ageMean)
df['Age']
```
