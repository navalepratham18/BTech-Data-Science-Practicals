Use NumPy/Pandas to calculate mean, median, mode, variance, std dev, correlation and covariance

```python
import numpy as np
import pandas as pd
from scipy import stats
```

## Mean

```python
data = [10,20,30, 40, 50]
avg1 = np.mean(data)    # numpy
print(avg1)

avg2 = pd.Series(data).mean()    #pandas
print(avg2)
```

## Median

```python
median_np = np.median(data)    #numpy
print(median_np)

median_pd = pd.Series(data).median()    #pandas
print(median_pd)
```

## Mode

```python
data = [1,2,3,3,3,3,5,6,2,4,4,4,4,5]
mode_val = stats.mode(data)
mode_val2 = pd.Series(data).mode()
print(mode_val)
print(mode_val2)
```

## Varience

```python
var_np = np.var(data)
var_pd = pd.Series(data).var()

print(var_np)
print(var_pd)
```

## Standard Deviation

```python
std_np = np.std(data)
std_pd = pd.Series(data).std()

print(std_np)
print(std_pd)
```
